## Abstract

Current protocol for managing residual waste in AguaClara plants consists of piping it away from the plant and releasing it directly to the environment. As Aguaclara plants increase in size and flow rate, alternative residual management solutions are sought to reduce cost and amount of water wasted. A lagoon system is an effective method to recycle wasted water back to the plant, and to dry and dispose of settled sludge. Such a process would require separate lagoons to settle backwash water and to dry sludge. Preliminary calculations were modeled after a 120 L/s plant in Gracias, Honduras. The backwash lagoon would need to hold around 780 $m^3$ of water and would require a velocity of 0.0180 mm/s. One model for the sludge lagoon requires an area between 290 and 600 m^2, depending on whether the climate is 'wet' or 'dry'. This corresponds to a velocity of 0.1405 mm/s. This area is highly variable, because it depends on the seasonal dry and wet period throughout the year. 

## Introduction
Residual waste from AguaClara plants consists of backwash water used to clean filters, sludge removed from the sedimentation basin, and wasted water during treatment failures. These materials by themselves do not present a threat to the environment, since they are primarily composed of clay, dirt, organic matter, and some coagulant. This waste is currently combined in a pipe, carried away from the plant, and released directly into the environment. With past AguaClara plants serving relatively small populations, this method of handling residuals has not provided an issue financially or environmentally. However, problems are arising as AguaClara plants continue to increase in size. The largest AguaClara plant to date was inaugurated on November 8, 2019 in the town of Gracias, Honduras. It serves a population of 20,000 people at a flow rate of 120 L/s. The town paid 1.7 million Lempiras (about 70,000 usd) just to build a residuals pipeline which is about 7% of the total plant cost. This cost could be reduced by designing on onsite residuals management system. Additionally, incorporating a recycle stream would reduce the amount of water wasted unnecessarily by AguaClara plants. 


### EPA Residual Stream Regulations

The EPA does not regulate the disposal of residual streams. However, other State and Federal regulations may apply in the United States [(*Filter Backwash Recycling Rule: Technical*, 2002)](https://nepis.epa.gov/Exe/ZyPDF.cgi?Dockey=200025V5.txt).

The EPA does provide minimal guidelines for recycling backwash water. The 2002 Filter Backwash Recycling Rule (FBRR) requires systems “to return their recycle flows through all processes of a system’s existing conventional or direct filtration system” [('Using DWSRF', 2002)](https://www.epa.gov/sites/production/files/2015-04/documents/fund-fbrr.pdf). These systems must be “recognized as capable of achieving 2-log (99 percent) Cryptosporidium removal" [(*Filter Backwash Recycling Rule: a Rule*, 2002)](https://nepis.epa.gov/Exe/ZyPDF.cgi?Dockey=200025UV.txt). No other requirements are currently in effect in the United States.

Exceptions to the Filter Backwash Recycling Rule are granted by the EPA if the following information is provided [('Using DWSRF', 2002)](https://www.epa.gov/sites/production/files/2015-04/documents/fund-fbrr.pdf):
- Evidence that a plant can handle hydraulic disruption caused by recycle stream addition to the influent
- diagrams depicting the "origin of all recycle flows, the hydraulic conveyance used to transport them, and the location where they are recycled back into the plant”
- plant design flow rate, average recycle flow, and "maximum plant flow experienced in the previous year" (all reported in gpm)
- plant operating capacity [if applicable]

Additionally, the EPA recommends thorough mixing of the recycle stream with raw influent water to minimize hydraulic disruption, and to implement a continuous recycle at a low flow rate rather than large batch additions [(*Filter Backwash Recycling Rule: a Rule*, 2002)](https://nepis.epa.gov/Exe/ZyPDF.cgi?Dockey=200025UV.txt).

A complicating factor in residual management design is the fact that residual wastes vary widely in turbidity and flow rate. Settled sludge from the floc hopper constitutes a very small percentage of the total residual waste flow, and is composed of highly concentrated flocs. In contrast, the majority of residual wastes come from spent filter backwash water, which has a very low turbidity compared to the sludge. This variability suggests that backwash water and wasted sludge should be treated separately and with different specifications.

There are two main ways to manage residual waste: 1) a miniature plant and 2) lagoons. A miniature plant could be used so that the backwash water and sludge would run through it. An additional electrical pump would need to be installed to recycle the water back up to the top of the original water treatment plant. This method would easily meet any and all of the EPA regulations and provide even cleaner water to rerun through the plant. However, building a miniature plant is more costly, presents fabrication challenges, requires regular monitoring, and would still produce residual wastes of its own. Furthermore, it does not solve the problem of getting rid of sludge and is more of a redundant process where the water just gets treated over again. There is an even cheaper alternative which is building a set of sludge lagoons and a set of backwash lagoons. A new AguaClara plant would be designed to keep backwash water and sludge separate so that they could enter their respective lagoon without mixing. This separation is ideal due to the large differences in turbidity and volume of the two wastes. Existing AguaClara plants could separate backwash water and sludge by laying a row of bricks in the residuals channel. A minimum of two sludge lagoons are needed so that one can fill up while the other is dried and emptied. Two backwash lagoons are also needed to handle variable flow. This method is much more cost-effective compared to building a miniature plant.

### Lagoon Structure and Materials

A lagoon consists of a large pit dug into the ground, lined with either concrete, clay or a synthetic liner [(*Technology*, 2001)](https://www3.epa.gov/npdes/pubs/cafo_cost_method_p3.pdf). The purpose of a lagoon can be either to settle out particles from water or to evaporate excess water from sludge, depending on the volume and turbidity of the residual. For a settling lagoon, water enters near the bottom/middle of the lagoon, and clarified water exits near the top to be recycled. Lagoons often contain emergency drain lines to handle flood risks, and are surrounded by a mound of excavated land called the embankment. For a sludge lagoon, particles build up cake-like layers as water evaporates. When the layers fill the lagoon, the sediment is dug out and disposed of manually or with heavy machinery such as a backhoe. Sludge lagoons are typically designed so that it will have to be dug out every 3-4 months once it reaches the freeboard height ("4.2 Sludge", pg. 395). The freeboard height is the distance between the water line and the top of the embankment. Figure 1 below shows an example of a lagoon and the pathway of water flowing in and out.

**Lagoon Materials**

Lagoons typically contain either a concrete, clay, or synthetic liner. The purpose of this liner is to limit groundwater contamination and to make sludge removal easier. It also suppresses weed growth and extends the longevity of the lagoon by protecting it from erosion (Kawamura, 2000). 
The EPA recommends using concrete, because it offers high levels of stability to the sloped sides of the lagoon structure, especially in locations with high sand content[(*Technology*,2001)](https://www3.epa.gov/npdes/pubs/cafo_cost_method_p3.pdf). Concrete is also sturdy when dealing with an influx of flow and provides extra strength to prevent any leakage contamination into the soil. Concrete’s main ingredient is portland cement and creating this cement results in high levels of $CO_2$ outputs, resulting in 4-5% of the worlds CO2 output[(*Greenspec*,2019)](http://www.greenspec.co.uk/building-design/environmental-impacts-of-concrete/).
Therefore, due to its large carbon footprint, concrete is not an ideal option for an AguaClara lagoon. 

The USEPA also suggests using synthetic liners such as HDPE and clay liners. Both are  appealing due to their lower cost and lower environmental impact. Clay liner; however, seems to be the more optimal choice. Synthetic liners are only $2.41 \space \frac{USD}{ft^2}$whereas clay liners are only $0.39 \space \frac{USD}{ft^2}$. Since the contents of backwash water and sludge are rather  non-toxic, concerns about leakage are minimal, and a clay liner will be more than sufficient to provide a boundary between the lagoon contents and the  soil underneath. 



## Proposed Sludge Management System

The proposed solution for managing settled sludge waste was a lagoon drying system. This system was appealing, because of its low cost and maintenance, lack of mechanical requirements, and ability to handle intermittent and variable flow. This last point was especially important since the frequency at which a floc hopper is emptied highly depends on plant location and weather events. 

Three models were created for calculating the minimum area of a sludge lagoon. Model numbers 2 and 3 can be found Appendix A.


### Sludge Lagoon Design and Calculations (Model No. 1 of 3)

The physical and economic feasibility of using a lagoon system for sludge management depends on the minimum plan-view area needed to dry the sludge as fast as it is produced. This area can be calculated by equating the average flow rate of water entering the lagoon to the net rate of water removal by evaporation . 

**Equation 1**

$$Q_{Lagoon} = A_{Lagoon}[R_{evap} -R_{precip}]$$

**Variables**

$Q_{Lagoon}$ = volumetric flow rate of water entering the lagoon

$A_{Lagoon}$ = area of the lagoon

$R_{evap}$ = local evaporation rate

$R_{evap}$ = local precipitation rate

<br/>

The flow rate of water entering the lagoon can be found by taking the ratio of the sludge production rate, and the concentration of sludge in water (Equation 2). Note that the average sludge concentration has not yet been quantified for AguaClara plants, and this value will be estimated as 10 g/L for calculations.

**Equation 2**

$$Q_{lagoon} =  \frac{R_{Sludge\space Production}}{C_{sludge}}$$

<br/>

The rate of sludge production in the floc hopper is a function of   coagulant type, coagulant dose, and the average turbidty entering the plant.

**Equation 3**

$$R_{Sludge\space Production} = Q_{plant} [Dose_{coag} \cdot {Ratio_{precipitate}} + Turbidity_{raw} \cdot Ratio_{TSS}]$$

**Variables:**

$Rate_{Sludge \space Producion} $ = average rate of solids produced    by mass (mg/s)

$Q_{plant}$ = design flow rate of the plant (L/s)

$Dose_{coag} $ = average dose of coagulant (mg/L)

$Ratio_{precipitate}$ = ratio of the mass of coagulant precipitate to the mass of granular coagulant

$Turbidity_{raw}$ = average turbidity of the raw water (ntu)

$Ratio_{TSS} $ = ratio of total suspended solids (TSS) (mg/L) to turbidity (ntu)

<br/>

*Note that values for the TSS : Turbidity ratio range from 1.0 to 2.0 and average around 1.3 (Kawamura, 2000, pg. 385). [AguaClara code](https://github.com/AguaClara/aguaclara/blob/master/aguaclara/core/units.py) supplies a ratio of 1.47 for the specific use of kaolinite clay and PACl.*

The precipitate ratio ($Ratio_{precipitate}$) depends on the type of coagulant used. For example, aluminum sulfate combines with calcium bicarbonate to produce solid Aluminum Hydroxide by Equation 2 below. 

**Equation 4**
$$Al_2(SO_4)_3\cdot14H_2O+3Ca(HCO_3)_2 \longrightarrow 3CaSO_4 + 2Al(OH)_3 + 6CO_2 + 14H_2O$$

The ratio of the mass of alum precipitate to the mass of granular alum can be calculated using this reaction stoichiometry as follows:

**Equation 5**
$$Ratio_{alum} = \frac{M_{precipitate}}{M_{granularAlum}}= \frac{2\cdot MW_{Al(OH)_3}}{MW_{Al_2(SO_4)_3\cdot14H_2O}}  = 0.26$$

A similar calculation may be performed for PACl via Equation 4. Note that the molecular formula for PACl depends on the degree of  polymerization and takes the general form of $[Al_n(OH)_mCl_{3n-m}]_x$  [(Gebbie, 2001)](http://wioa.org.au/conference_papers/2001/pdf/paper6.pdf).

Where
$1 \leq n \leq 5$ and $m\leq  10$ [('Poly')](https://www.takichem.co.jp/eng/products/water/pac.html)

**Equation 6**
$$Ratio_{PACl} = \frac{M_{precipitate}}{M_{granularPACl}}= \frac{n\cdot MW_{Al(OH)_3}}{n\cdot MW_{Al} + m\cdot MW_{OH} + (3n-m)\cdot  MW_{Cl}} $$  

<br/>

Equations 1-3 may be combined to form a single formula for the minimum area of a sludge lagoon, based on the design flow rate of the plant, average coagulant dose, type of coagulant used, turbidity of the raw water, and climate. Corresponding python calculations can be found in the [Code and Calculations](#Code-and-Calculations) section.

**Equation  7**

$$A_{Lagoon} = \frac{Q_{lagoon}}{[R_{evap} - R_{precip}]} = \frac{Q_{plant}[Dose_{coag} \cdot {Ratio_{precipitate}} + Turbidity_{raw} \cdot Ratio_{TSS}]}{C_{sludge}[R_{evap} -R_{precip}]}$$

**Variables:**

$Q_{plant}$ = design flow rate of the plant (L/s)

$Dose_{coag} $ = average dose of coagulant (mg/L)

$Ratio_{precipitate}$ = ratio of the mass of coagulant precipitate to the mass of granular coagulant

$Turbidity_{raw}$ = average turbidity of the raw water (ntu)

$Ratio_{TSS} $ = ratio of total suspended solids (TSS) (mg/L) to turbidity (ntu)

$C_{Sludge}$ = the average concentration of sludge in water from the floc hopper (mg/L)

$R_{evap}$ = annual average evaporation rate (typically reported in mm//month)

$R_{precip}$ = annual average precipitation rate (typically reported in mm/yr)

<br/>

Since the lagoon is designed such that water evaporates as fast as it enters the lagoon (averaged over a long period of time), the depth of the lagoon is solely a function of how often the operator wishes to excavated and removed dried solids. A deeper lagoon may cost more up front to build, but would require less frequent maintenance. Kawamura suggests an ideal height of wet sludge filling the lagoon to be 4 ft, with a maximum of 6 ft allowed (2000, pg. 395). These requirements suggest an approximate lagoon depth of 7-8 ft to allow for flexibility in operation and to prevent flooding in a high precipitation weather event. A total of 2 lagoons should be built to handle sludge waste, so that one can fill while the other is dried and emptied. 

The shape of the lagoons should be of an inverted frustum pyramid, as seen in Figure 2 below. The area calculated in equation 7 above refers to the area of the smaller base. The sides should have a slope of 3:1 and the length:width ratio should be 4:1 for the bottom and top base. Having a sloped side geometry enables front-end-loader access to remove solids and helps prevent erosion from sludge and backwash when entering the lagoon [(*Technology*, 2001)](https://www3.epa.gov/npdes/pubs/cafo_cost_method_p3.pdf).

In order to compare the size of the lagoon with other treatment processes, the results can be written as a velocity as follows:

**Equation 8**

$$V = \frac{Q_{plant}}{A_{Lagoon}}$$

However, the area of interest is the wider, upper base of the lagoon and not the area calculated previously. The area of the upper base can be calculated with some simple geometry.

Given the 4:1 side ratio mentioned previously, and the area of the bottom base, $A_B$, the bottom width is

**Equations  9 - 12**

$$ W_B = \sqrt{\frac{A_B}{4}}$$

<center>

The length of the bottom base is

$$ L_B = 4W_B$$

Knowing that the sloped sides horizontal:vertical is 3:1. The width of the top base is 

$$W_T = W_B + 6H$$

The length of the top base is

$$L_T = L_B + 6H$$

and therefore the top area is 

</center>

**Equation 13**

$$A_T = L_T \cdot W_T = (L_B + 6H)(W_B + 6H) = A_B +  3h\sqrt{A_B} + 36H^2$$

Therefore the velocity needed for the sludge lagoon is given by Equation 14 below:

**Equation  14**

$$V = \frac{Q_{plant}}{A_{Lagoon}} = \frac{Q_{plant}}{A_B +  3H\sqrt{A_B} + 36H^2}$$

Where $A_B$ is the area of the lagoon bottom and H is the height of the lagoon. 

## Proposed Backwash Management System
 
The EPA does not require, although it does recommend, that the backwash water go through any sort of treatment process in order to be recycled (*Filter Backwash Recycling Rule: Technical*, 2002). Because of this, there is no need for any of the AguaClara plants to send the water from backwashing the filters anywhere else except for back up to the beginning of the plant. However, each filter will get backwashed on average once per day for fifteen minutes. Therefore, there will be times when there is no water from backwash entering the plant and then times when there is a high influx of this water. The backwashing water will be higher in turbidity than the raw water coming into the plant on average. It will be hard for the operators to judge how much coagulant to use and how to treat the water. To solve this we can design to keep the backwashing water in some sort of storage tank. Once the backwashing water is in these tanks we can feed a small steady stream of this backwashing water as to not interrupt the turbidity significantly.
 
Even though there is no need to have any of the particles settle, since AguaClara plants will require a storage of this water, designing this tank to be a lagoon that will settle out at least some of the particles is the most efficient and cost saving option. There will already need to be a tank, which has a cost, and removing any, even if it is a small amount, of particles will also require less treatment of that water since its turbidity will be lowered. Therefore, the backwash water will be fed into a lagoon. The lagoon design will have the same constraints of any of the other lagoons. Thus, its depth will be 3m and the ratio of length to width will be 4:1 and a slope of 3:1 on the sides. This will be the necessary requirments to let any particles settle if they can. Since these particles will already be treated with coagulent from the plant, there is a good chance that they won't dirupt turbidity too much (Kawamura, 2000, pg. 377). 
 
The backwash water will need to get back up to the top of the plant since it will be at a lower elevation once it is done backwashing. In this case an electric pump will be needed. Once the filters are backwashed, the water will feed into the lagoon. There will be a pump that can give a constant flow of 3-5 percent of the plant flow. However, this pump will be inefficient, since this flow would be around 1 L/s. For the purposes of an AguaClara plant this is acceptable because any of the water that can be recycled is a positive. 
 
The backwashing lagoon will be a set of two, as the EPA recommends. Because we cannot predict the amount of backwashing needed during a large rainstorm event, having a set of two will allow for any major rainfall event to be captured. The lagoons will have drains as well so that if a unprecedented event happens the water can be drained. 

### Backwash Lagoon Design and Calculations

The size of the each lagoon should be big enough to handle a backwash from all filters in a single day. Additionally, when many backwashes are needed in a given day there is likely a large rain event so a safety factor will be included to account for the lagoons filling up that way as well. 

First, the flow through a filter is calculated using the minimum velocity needed to fluidize the bed and multiplying it by the cross-sectional area of the bed:
$$v_{MinFluidization} = \frac{\phi_{FiSand}^3 g D_{60}^2}{36 k \nu \left( 1 - \phi_{FiSand} \right)} \left( \frac{\rho_{Sand}}{\rho_{Water}} - 1 \right)$$
The crosssectional area of the filter will be different based on each plants already calculated needs, so that will just be an input in calculating the flow rate of the backwash.
$$Q_{Backwash} = v_{MinFluidization} * A_{x-section\space bed}$$
Then to transform this into a volume of water the flow will be multiplied by the backwashing time. This input is also at the whim of the plant and operators at each plant. So, the time will also be an input:
$$V_{Backwash} = Q_{Backwash} * T_{Backwash}$$
This is for one bed so this will need to be multiplied by the number of beds in the plant to get a volume of water for one backwash per filter in a day:
$$TotalV_{Backwash} = V_{Backwash} * N$$
where $N$ = Number Beds

Each of the lagoons should be big enough to hold a days worth of backwashing under normal condition. The lagoon will also factor in a safety factor of 20% more backwashing per day, in the case of larger rainfall events:
$$V_{Lagoon} = TotalV_{Backwash} * (1.2)$$
Depth of 3 meters on each lagoon and a length to width of 4:1:
$$SA_{Lagoon} = \frac{V_{Lagoon}}{3_{meters}}$$
$$Length_{Lagoon} = 4 * Width_{Lagoon}$$
$$Width_{Lagoon} = (\frac{SA_{Lagoon}}{4})^{\frac{1}{2}}$$


###Constraints to Backwash Lagoon
The amount of water in each backwashing event will need to fit in the lagoon. Additionally, the lagoon's geometry will depend on the best fit so that the any particles left in the water will have some sort of chance to settle out.

The pump will pump the backwash water up to the inlet of the plant from the lagoons. This pump will be running at all times to ensure a steady flow of 1-3% of the total flow of the plant. Since, normal operating conditions are to backwash for 15 minutes and adding the safety factor becomes 18 minutes, the total flow of backwashing in one day should be 18 out of the total 1440 minutes in a day.

Only the 2nd model produced an area that was conceivable reasonable for an AguaClara plant to implement. On the mountainous terrain of Honduras, it may be difficult to find a large section of land suitable for excavating such large basins, and the shear size predicted   in models 1 and 3 would make the project cost prohibitive. The cost of building two sludge lagoons with base areas of  541 $m^2$ was found to be 27,000 USD with annual operating costs of 674.80 USD each. This installation cost is less than 40% of the total pipeline cost spent at the plant in Gracias, Honduras to remove residual waste. Therefore, lagoons are an extremely cost effective way to handle residual sludge if something can be done to decrease the required land area. The biggest contribution to land area in models 1 and 2 was the low deficit between evaporation rates and precipitation rates estimated for Honduras. Additionally, the precipitation rate for regions of Honduras may have been underestimated, meaning that the actual lagoon area needed is greater than those calculated in this report. An important failure mode for the design of a sludge lagoon occurs when the average precipitation rate in a region is greater than its evaporation rate. This net accumulation of water in the lagoon would preclude sludge drying. One possible way to mitigate the effect of precipitation on sludge drying rates would be to ensure a broad and tall embankment surrounded by sufficient plant coverage to prevent flooding and erosion. Additionally, the sludge lagoon could be fitted with a drainage system that would carry excess water to the backwash lagoon. Doing so would minimize the amount of water that needs to be removed by evaporation.

### Sludge Lagoon Design and Calculations (Model No. 2 of 3)

The physical and economic feasibility of using a lagoon system for sludge management depends on the minimum plan-view area needed to dry the sludge as fast as it is produced. The following equations calculating this area were adapted from Kawamura's empirical formula for the production rate of dry alum sludge (2000, pg. 385). Corresponding python code can be found in the [Code and Calculations](#Code-and-Calculations) section:

**Equation 1:**

$$Rate_{sludgeProduction} = Q_{plant} [Dose_{coag} \cdot {Ratio_{precipitate}} + Turbidity_{raw} \cdot Ratio_{TSS}]  $$

**Variables:**

$Q_{plant}$ = design flow rate of the plant (L/s)

$Rate_{sludgeProduction} $ = average rate of sludge produced (mg/s)

$Ratio_{precipitate}$ = ratio of the mass of coagulant precipitate to the mass of granular coagulant

$Dose_{coag} $ = average dose of coagulant (mg/L)
 
$Turbidity_{raw}$ = average turbidity of the raw water (ntu)

$Ratio_{TSS} $ = ratio of total suspended solids (TSS) (mg/L) to turbidity (ntu)

<br/>

*Note that values for the TSS : Turbidity ratio range from 1.0 to 2.0 and average around 1.3 (Kawamura, 2000, pg. 385). [AguaClara code](https://github.com/AguaClara/aguaclara/blob/master/aguaclara/core/units.py) supplies a ratio of 1.47 for the specific use of kaolinite clay and PACl.*

The precipitate ratio ($Ratio_{precipitate}$) depends on the type of coagulant used. For example, aluminum sulfate combines with calcium bicarbonate to produce solid Aluminum Hydroxide by Equation 2 below. 

**Equation 2**
$$Al_2(SO_4)_3\cdot14H_2O+3Ca(HCO_3)_2 \longrightarrow 3CaSO_4 + 2Al(OH)_3 + 6CO_2 + 14H_2O$$

The ratio of the mass of alum precipitate to the mass of granular alum can be calculated using this reaction stoichiometry as follows:

**Equation 3**
$$Ratio_{alum} = \frac{M_{precipitate}}{M_{granularAlum}}= \frac{2\cdot MW_{Al(OH)_3}}{MW_{Al_2(SO_4)_3\cdot14H_2O}}  = 0.26$$

A similar calculation may be performed for PACl via Equation 4. Note that the molecular formula for PACl depends on the degree of  polymerization and takes the general form of $[Al_n(OH)_mCl_{3n-m}]_x$  [(Gebbie, 2001)](http://wioa.org.au/conference_papers/2001/pdf/paper6.pdf).

Where
$1 \leq n \leq 5$ and $m\leq  10$ [('Poly')](https://www.takichem.co.jp/eng/products/water/pac.html)

**Equation 4**
$$Ratio_{PACl} = \frac{M_{precipitate}}{M_{granularPACl}}= \frac{n\cdot MW_{Al(OH)_3}}{n\cdot MW_{Al} + m\cdot MW_{OH} + (3n-m)\cdot  MW_{Cl}} $$

<br/>

Kawamura also presents an empirical conversion factor relating sludge production and the plan-view area needed for the lagoon (2000, pg. 395). The area needed for the lagoon depends on the mass of sludge produced over a period of 3-4 months, and whether the climate is ‘wet’ or ‘dry’. A timeframe of 3.5 months will be used for further calculations. 

In a wet region, the square footage needed per pound of dry sludge is: 

$$area_{wet} =  2.56 \cdot 10^{-8} m^2/mg$$

In a dry region, the square footage needed per pound of sludge is:

$$area_{dry} = 1.28\cdot 10^{-8} m^2/mg$$

Combining the rate of sludge production and this conversion ratio produces an equation for lagoon size:

**Equation 5:**

$$Area_{lagoon} = Q_{plant} \cdot Time_{accumulation} \cdot  R_{sludgeProduction} \cdot  Area_{conversion}$$ 

**Variables:**

$Q_{plant} $ = design flow rate of the plant 

$Time_{accumulation} $ = 3.5 months as described above

$R_{sludgeProduction} $ = rate of sludge production via Eqn.1
 
$Area_{conversion} $ = either $area_{dry}$ or $area_{wet}$

<br/>

Equation 1 and Equation 5 may be combined to form a single formula for the minimum area of a sludge lagoon based on the design flow rate of the plant, raw water quality, coagulant choice, and average coagulant dose. 

**Equation 6**
$$Area_{lagoon} = Q_{plant}\cdot  Time_{accumulation}\cdot [Dose_{alum}\cdot Ratio_{precipitate} + Turbidity_{raw}\cdot  Ratio_{TSS}] \cdot Area_{conversion}$$

**Variables:**

$Q_{plant} $ = design flow rate of the plant 

$Time_{accumulation} $ = 3.5 months as described above

$Dose_{alum} $ = average dose of alum coagulant (mg/L)

$Ratio_{precipitate}$ = ratio of the mass of coagulant precipitate to the mass of granular coagulant
 
$Turbidity_{raw}$ = average turbidity of the raw water (ntu)
  
$Ratio_{TSS} $ = ratio of total suspended solids (mg/L) to turbidity (ntu)
 
$Area_{conversion} $ = either $area_{dry}$ or $area_{wet}$





```
# Sludge Lagoon Calculations (Model N0. 2 of 3)

# Calculates the mass ratio of Aluminum Hydroxide Precipitate to Granular PACL in the form of [Al_n(OH)_mCl_(3n-m)]_x
def Ratio_PACl(n,m):
    MW_precip = 78 * u.g/u.mol
    MW_Al = 26.98 * u.g/u.mol
    MW_OH = 17.01 * u.g/u.mol
    MW_Cl = 35.45 * u.g/u.mol
    Ratio = n*MW_precip/(n*MW_Al + m*MW_OH + (3*n-m)*MW_Cl)
    return Ratio

# Calculate the minimum area needed for a lagoon such that water evaporates as fast as it enters the lagoon
Q_plant = 120 * u.L/u.s
Dose_coag = 10 * u.mg/u.L# mg/L an arbitrary PACl dosage taken from the FCM design challenge (will consider changing this later)
Turbidity_raw = 8.0 * u.NTU
Ratio_precip = Ratio_PACl(2,3)
R_sludge_prod = Dose_coag*Ratio_precip + Turbidity_raw.to(u.mg/u.L)
 
Time_accum = (3.5 * u.month).to(u.s) # the span over which sludge is accumulated
galH20_treated = (Q_plant * Time_accum).to(u.gal)
lb_sludge = galH20_treated * R_sludge_prod
Area_wet = ((1/8) * u.ft**2/u.lb) *  lb_sludge   # plan view area needed in wet regions
Area_dry = ((1/16) * u.ft**2/u.lb) *  lb_sludge  # plan view area needed in dry regions

print('The area needed for a sludge lagoon in a wet climate is ', (Area_wet).to(u.m**2))
print('The area needed for a sludge lagoon in a dry climate is ', (Area_dry).to(u.m**2))

# Calculate the velocity needed for this lagoon
H = (7 * u.ft).to(u.m)
A = (Area_wet).to(u.m**2) # wet version chosen arbitrarily
A_upper = A + 3*H*(A)**0.5 + 36*H**2 
V_needed = (Q_plant/A_upper).to(u.m/u.s)
print('The velocity needed for this lagoon is ', V_needed)
```

### Sludge Lagoon Design and Calculations (Model No. 3 of 3)

The physical and economic feasibility of using a lagoon system for sludge management depends on the minimum plan-view area needed to dry the sludge as fast as it is produced. This area can be calculated by considering the total amount of water that needs to evaporate and how quickly that will happen in a given climate, by Equation 1 below. 

**Equation 1:**

$$Area \space Needed = \frac{V_{excess \space water}}{[R_{evap} - R_{Precip}]\theta}$$

**Variables**

$V_{excess \space water}$ = the amount of water that needs to evaporate before the sludge is considered dry and is ready for removal

$R_{evap}$ = annual average evaporation rate [m/month]

$R_{precip}$ = annual average precipitation rate [m/month]

$\theta$ = turnover time, or the time by which the sludge bed is expected to be dry and solids are dug out [months] [arbitrarily chosen to be 3.5 months based on Kawamura's work (2000, pg.395)]

The excess water that needs to be evaporated is the difference in water content of 'wet' and 'dry' sludge (Equation 2), which is characterized by porosity (Equation 3). 

**Equation 2**

$$V_{excess \space water} = V_{excess \space wet} - V_{excess \space 
dry}$$


**Equation 3**

$$\rho = \frac{V_{water}}{V_{water} + V_{solids}}$$

Equations 4 and 5 below show how the porosity equation can be rearranged to describe $V_{excess \space wet}$ and $V_{excess \space 
dry}$, respectively.

**Equation 4**

$$V_{excess \space wet} = \frac{-\rho_d V_s}{\rho_d - 1}$$

**Equation 5**

$$V_{excess \space dry} = \frac{-\rho_w V_s}{\rho_w - 1}$$

**Variables**

$V_s$ = volume of solids produced in turnover time $\theta$ [$m^3$]

$\rho_d$ = the porosity of 'dry' sludge, approximated as 0.40 

$\rho_w$ = the porosity of 'wet' sludge calculated in Equation 6 below

**Equation 6**

$$\rho_{wet} = \frac{Q_{Sludge \space Water} - Q_{Solids}}{Q_{Sludge \space Water}}$$

The volumetric flow rate of sludge water flowing to the lagoon can be calculated as a  fraction, *f*, of total plant flow (Equation 7). Assuming backwash wastes between 3 and 5% of total plant flow, water diverted to the  sludge lagoon was estimated to be 0.5% of total plant flow (ie. *f* = 0.005).

**Equation 7**

$$Q_{Sludge \space Water} = f \cdot Q_{plant}$$

The volumetric flow rate of solids and the total volume of solids produced in turnover time $\theta$ can be calculated with Equations 8 and 9, respectively, where $D_{Solids}$ is the density of solid particles in $kg/m^3$.

**Equation 8**

$$Q_{Solids} = \frac{R_{Sludge \space Production}}{D_{Solids}}$$

**Equation 9**

$$V_s = \frac{R_{Sludge \space Production}\theta}{D_{Solids}}$$

The rate of sludge production was adapted from Kawamura's empirical formula for the production rate of dry alum sludge (2000, pg. 385).

$$Rate_{sludgeProduction} = Q_{plant} [Dose_{coag} \cdot {Ratio_{precipitate}} + Turbidity_{raw} \cdot Ratio_{TSS}]$$

**Variables:**

$Q_{plant}$ = design flow rate of the plant (L/s)

$Rate_{sludgeProduction} $ = average rate of sludge produced (mg/s)

$Ratio_{precipitate}$ = ratio of the mass of coagulant precipitate to the mass of granular coagulant

$Dose_{coag} $ = average dose of coagulant (mg/L)
 
$Turbidity_{raw}$ = average turbidity of the raw water (ntu)

$Ratio_{TSS} $ = ratio of total suspended solids (TSS) (mg/L) to turbidity (ntu)

*Note that values for the TSS : Turbidity ratio range from 1.0 to 2.0 and average around 1.3 (Kawamura, 2000, pg. 385). [AguaClara code](https://github.com/AguaClara/aguaclara/blob/master/aguaclara/core/units.py) supplies a ratio of 1.47 for the specific use of kaolinite clay and PACl.*

The precipitate ratio ($Ratio_{precipitate}$) depends on the type of coagulant used. For example, aluminum sulfate combines with calcium bicarbonate to produce solid Aluminum Hydroxide by Equation 2 below. 

**Equation 10**
$$Al_2(SO_4)_3\cdot14H_2O+3Ca(HCO_3)_2 \longrightarrow 3CaSO_4 + 2Al(OH)_3 + 6CO_2 + 14H_2O$$

The ratio of the mass of alum precipitate to the mass of granular alum can be calculated using this reaction stoichiometry as follows:

**Equation 11**
$$Ratio_{alum} = \frac{M_{precipitate}}{M_{granularAlum}}= \frac{2\cdot MW_{Al(OH)_3}}{MW_{Al_2(SO_4)_3\cdot14H_2O}}  = 0.26$$

Where
$1 \leq n \leq 5$ and $m\leq  10$ [('Poly')](https://www.takichem.co.jp/eng/products/water/pac.html)

**Equation 12**
$$Ratio_{PACl} = \frac{M_{precipitate}}{M_{granularPACl}}= \frac{n\cdot MW_{Al(OH)_3}}{n\cdot MW_{Al} + m\cdot MW_{OH} + (3n-m)\cdot  MW_{Cl}} $$

<br/>

Equations 1 -  9  may be combined to form a single formula for the minimum area of a sludge lagoon based on the design flow rate of the plant, fraction of influent water wasted to sludge removal, raw water quality, coagulant choice, average coagulant dose, and climate. Corresponding python calculations can be found in the [Code and Calculations](#Code-and-Calculations) section.

**Equation 13**

$$A = \frac{Q_{plant} [Dose_{coag} \cdot {Ratio_{precipitate}} + Turbidity_{raw} \cdot Ratio_{TSS}]}{D_{Solids}[R_{evap}-R_{Precip}]} \cdot [\frac{\rho_d}{\rho_d - 1} + \frac{f \cdot D_{Solids}}{[Dose_{coag} \cdot {Ratio_{precipitate}} + Turbidity_{raw} \cdot Ratio_{TSS}]} - 1]$$

**Variables:**

$Dose_{coag} $ = average dose of coagulant (mg/L)

$Ratio_{precipitate}$ = ratio of the mass of coagulant precipitate to the mass of granular coagulant

$D_{Solids}$ = the density of solid particles ($kg/m^3$).

$R_{evap}$ = annual average evaporation rate (m/month)

$R_{precip}$ = annual average precipitation rate (m/month)

```
# Sludge Lagoon Calculations (Model No. 3 of 3)

# Calculate the mass ratio of Aluminum Hydroxide Precipitate to Granular PACL in the form of [Al_n(OH)_mCl_(3n-m)]_x.

def Ratio_PACl(n,m):
    MW_precip = 78 * u.g/u.mol
    MW_Al = 26.98 * u.g/u.mol
    MW_OH = 17.01 * u.g/u.mol
    MW_Cl = 35.45 * u.g/u.mol    
    Ratio = n*MW_precip/(n*MW_Al + m*MW_OH + (3*n-m)*MW_Cl)
    return Ratio

# Calculate the minimum area needed for a lagoon such that water evaporates as fast as it enters the lagoon
Q_plant = (120 * u.L/u.s).to(u.m**3/u.s) # design flow  rate  for Gracias, Honduras plant
Dose_coag = (10 * u.mg/u.L).to(u.kg/u.m**3) # an arbitrary PACl dosage
Ratio_precip = Ratio_PACl(2,3)
Turb_raw = 8.0 * u.NTU 
R_evap = (61 * u.mm/u.month).to(u.m/u.s)
R_precip = (300 * u.mm/u.year).to(u.m/u.s)
rho_d = 0.40  
D_solid = 1400 * u.kg/u.m**3
f = 0.005
R_sp = Q_plant*(Dose_coag*Ratio_precip + Turb_raw.to(u.kg/u.m**3))
A = (R_sp/(D_solid*(R_evap - R_precip)))*(rho_d/(rho_d-1) + D_solid*f*Q_plant/R_sp - 1)
print('The minimum area needed for the sludge lagoon is ', A)

# Calculate the velocity needed for this lagoon
H = (7 * u.ft).to(u.m)
A_upper = A + 3*H*(A)**0.5 + 36*H**2 
V_needed = Q_plant/A_upper
print('The velocity needed for this lagoon is ', V_needed)

```
