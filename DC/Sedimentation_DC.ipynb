## Design thinking
Both the diffuser and the plate settler designs are somewhat arbitrary. The diameter of the diffusers has been set to 1" nominal diameter. The spacing of the plate settlers has also been set to 1" (2.5 cm). You should be very skeptical about whether those are actually optimal designs!
